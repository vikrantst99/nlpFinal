### The three major blocks:-
- Block 1:- (TOOLS)
    - The searcher. (Searches the FAISS vector store and returns relevant chunks (Their embeddings from FAISS))
    - The getter. (Given the embeddings of a chunk, returns the metadata)
- Block 2:- (The Brain)
    - This is Qwen3 8B running through Ollama. In LangChain, we connect to it using ChatOllama. The LLM acts as the brain — it reads tool results and decides what to do next.
- Block 3:- The Agent Executor
    - This is the LangChain component that runs the loop:
        - LLM thinks → calls a tool → reads result → LLM thinks again → calls a tool or stops

- GOAL        → User asks a strategic question
- PLAN        → Agent decides what to search for
- RETRIEVE    → Agent calls search_nvidia_knowledge tool
- ANALYZE     → Agent reads chunks and identifies patterns
- DECIDE      → Agent determines if more retrieval is needed
- RECOMMEND   → Agent formulates a strategic recommendation
- VALIDATE    → Agent checks: do I have evidence? Are sources cited?
- OUTPUT      → Final structured answer with evidence

#### CRAM THE FOLLOWING AND NEVER EVER FORGET THEM :- 
- ChatOllama — connects to your local Qwen3 8B
- @tool decorator — turns a Python function into an agent tool
- create_tool_calling_agent — builds the agent logic
- AgentExecutor — runs the loop
- FAISS.load_local() — loads your vector store
- .as_retriever() — connects FAISS to the agent
