```python
from langchain_core.documents import Document
from langchain_community.retrievers import BM25Retriever
import pandas as pd
```

```python
df=pd.read_csv("../../data/chunked/langChainchunks.csv")
```

```python
em=[]
for i, j in df.iterrows():
    doc=Document(page_content=j["chunk_text"], metadata={"source":j["source"], "URL":j["url"], "title":j["title"]})
    em.append(doc)
```

```python
BM25R=BM25Retriever.from_documents(em)
BM25R.k=2
```

```python
query="gaming"
result=BM25R.invoke(query)
```

```python
for i, j in enumerate(result):
    print(f"At position {i+1} is : ")
    print(j.page_content[:1500])
```
